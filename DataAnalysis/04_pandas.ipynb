### Series
- 인덱스(라벨)가 붙은 1차원 배열
- 인덱스 이름을 직접 지정할 수 있음

```python
# 시리즈 생성
# 1. 리스트로 생성
data = [10,20,30,40]
s = pd.Series(data)
print(s)
```

```python
# 2. 인덱스를 직접 지정
s = pd.Series(data, index=["a", "b", "c", "d"])
print(s)
```

```python
# 3. 딕셔너리로 생성
data = {'서울': 100, '부산': 200, '대구': 300}
s = pd.Series(data)
print(s)
```

```python
# 4. 하나의 값으로 시리즈 생성
s = pd.Series(7, index = ["가", "나", "다"])
print(s)
```

```python
# 시리즈의 주요 속성
data = [10,20,30,40]
s = pd.Series(data, index = ['a', 'b', 'c', 'd'], name = 'test data')

print("values:", s.values) # 실제 데이터 배열
print("index:", s.index) # 인덱스 개체
print("dtype:", s.dtype) # 실제 데이터의 타입
print("size:", s.size) # 데이터 개수
print("shape:", s.shape) # 배열의 형태
print("name:", s.name) # 시리즈 이름
 
```

```python
# 시리즈의 연산
s1 = pd.Series([10,20,30], index = ['a', 'b', 'c'])
s2 = pd.Series([5,6,7], index = ['c', 'b', 'd'])

result = s1 + s2
print(result)
```

```python
# 브로드 캐스팅
s = pd.Series([5, 10, 15])

print(s + 3)
print(s * 2)
```

```python
# 불리언 연산 및 인덱싱
s = pd.Series([50,60,70,80,90,100])

mask = s > 70
print(mask)
print(s[mask])
```

### DataFrame
- 2차원 라벨 데이터 구조
- 시리즈를 모아놓은 것
- 행과 열이 있고, 행과 열에 대해 라벨을 부여할 수 있음

```python
# DataFrame 생성
# 1. 딕녀너리로 생성

data = {
    "이름" : ["kim", "lee", "sin"],
    "나이" : [15, 25,33],
    "도시" : ["서울", "부산", "대구"]
}

df = pd.DataFrame(data)
df
```

```python
# 2. 리스트의 리스트(2차원 리스트)로 생성

data = [[1, "a"], [2, "b"], [3, "c"]]
df = pd.DataFrame(data, columns = ["번호", "코드"])
df
```

```python
# 3. 딕셔너리의 리스트 생성

data = [
    {"이름": "kim", "나이": 20, "사는 곳": "서울"},
    {"이름": "choi", "나이": 30, "사는 곳": "대구"}
]

df = pd.DataFrame(data, index=["a", "b"])
print(df)
```

```python
# 4 시리즈의 딕셔너리로 생성

s1 = pd.Series(["국어", "수학", "영어"], index = ["a", "b", "c"])
s2 = pd.Series([100, 90, 85], index = ["a", "b", "c"])

df = pd.DataFrame({"과목": s1, "점수": s2})
df
```

```python
# 데이터 프레임 기본 속성

data = [
    {"이름": "kim", "나이": 20, "사는 곳": "서울"},
    {"이름": "choi", "나이": 30, "사는 곳": "대구"}
]

df = pd.DataFrame(data)

print("shape:", df.shape)
print("columns:", df.columns)
print("index:", df.index)
print("dtypes:", df.dtypes)
print("values:", df.values)
```

### 인덱싱과 슬라이싱

```python
# 인덱싱 : 컬럼에 대해 인덱싱 적용
# 특정 컬럼(시리즈) 선택

df['이름']

# 여러 컬럼을 리스트로 선택
df[['이름', '나이']]

# .으로도 조회가 가능함
df.이름
```

```python
# 슬라이싱

df[1:4]
df[-3:]

# 슬라이싱은 행 기준으로 작동, 열 기준은 별도의 방법을 사용
# df[:, 0:2] # KeyError 
```

### iloc, loc

```python
# iloc
# Integer Location
# 정수 위치 기반의 인덱싱, 슬라이싱
# numpy의 슬라이싱과 거의 유사함

# 단일 행/열 선택

df.iloc[0] # 첫 번째 행
df.iloc[:, 1] # 두 번째 열
```

```python
# 여러 행/열 동시 선택

df.iloc[1:4]
df.iloc[:, 0:2]
df.iloc[1:4, 0:2]
```

```python
# Fancy Indexing

df.iloc[[0,2,4]]
df.iloc[:, [1,2]]
df.iloc[[0,2,4], [1,2]]
```

```python
# 음수 인덱스 슬라이싱

df.iloc[-1]
df.iloc[:, -2:]
```

```python
# loc
# location의 약자
# 라벨의 이름 기준으로 인덱싱/슬라이싱
# 단일 행/열 선택
# 시작과 끝을 모두 포함
# 음수 인덱스 사용x

df.loc[0]
df.loc[:, "이름"]
```

```python
# 여러 행/열 선택
df.loc[2:4, ["이름", "직업"]]
df.loc[2:4, "이름": "직업"]
```

```python
# 조건식
mask = df["나이"] >= 30
df.loc[mask, ["이름", "나이"]]
```

```python
# 컬럼을 인덱스로 지정
df2 = df.set_index("이름")
df2

df2.loc["이순신"]
```
