## 07. matplotlib
- 파이썬에서 가장 널리 사용되는 데이터 시각화 라이브러리!

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

matplotlib.__version__

plt.rcParams['font.family'] = 'Malgun Gothic'  # '맑은 고딕'이 설치되어 있을 경우
plt.rcParams['axes.unicode_minus'] = False     # 마이너스(-) 부호 깨짐 방지
```

### 그래프 그리기

```python
x = [1,2,3,4,5]
y = [10,13,17,19,24]

# 선 그래프
plt.plot(x, y)

# 콘솔에 이미지 그리기
plt.savefig("my_graph.png")

# 이미지를 파일로 저장
plt.show()
```

### 그래프 구성요소 (1)
- figure: 여러 그래프가 그려지는 최상위 컨테이너
- axes: 실제 그래프가 그려지는 공간
- axis: 그래프의 축

```python
# subplots()
# - 여러 개의 그래프 그릴 때 사용
# - figure, axes 동시 반환
# fig, axs = plt.subplots(행, 열)
fig, axs = plt.subplots() # 하나의 figure와 하나의 axes 만듦
```

```python
# 1행 2열의 axes 생성
fig, axs = plt.subplots(1, 2)

axs[0].plot([1,2,3], [1,2,3])
axs[1].plot([1,2,3], [1,4,9])

plt.show()
```

```python
# 2행 2열의 axes 생성
fig, axs = plt.subplots(2, 2)
a = [1,2,3]
b = [1,4,9]

fig.suptitle("여러 함수 그래프")

axs[0, 0].plot(a, a)
axs[0, 0].set_title("1번 그래프")

axs[0, 1].plot(a, b)
axs[0, 1].set_title("2번 그래프")

axs[1, 0].plot(a, [3,2,1])
axs[1, 0].set_title("3번 그래프")

axs[1, 1].plot(a, [9,4,1])
axs[1, 1].set_title("4번 그래프")

plt.tight_layout() # 그래프 간격 자동 조정
plt.show()
```

### 그래프의 구성요소 (2)
- 제목, 레이블, 범례, 그리드

```python
# title
plt.plot([1,2,3,4,5], [1,5,9,16,25], label = '제곱 그래프', linestyle = '--', color = "c", marker = "o")
plt.title("제곱 함수 그래프", color = "b", fontsize = 20)

# xlabel, ylabel
plt.xlabel("x 값")
plt.ylabel("y 값(x의 제곱)")

# 범례 (legend)
plt.legend()

# 격자 (grid)
plt.grid(linestyle = "--")

plt.show()
```

```python
# 하나의 axes에 여러개의 그래프 그리기
x = np.arange(0,10)
y1 = 2 * x + 1
y2 = 3 * x - 5

plt.plot(x, y1, marker = "o", label = "y = 2x + 1", color = "c")
plt.plot(x, y2, marker = "*", label = "y = 3x - 5", color = "m")

plt.title("두 개의 선 그래프 비교")
plt.xlabel("x값")
plt.ylabel("y값")
plt.legend()
plt.grid(linestyle = ":", alpha = 0.5)

plt.show()
```

#### 실습 1. 선 그래프 그리기

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
sales_2019 = [100,120,140,110,130,150,160,170,180,200,190,210]
sales_2020 = [90,110,130,120,140,160,170,160,150,180,200,190]

plt.plot(months, sales_2019, label = "2019")
plt.plot(months, sales_2020, label = '2020')

plt.title("Monthly Sales Comparison (2019-2020)")
plt.xlabel("Month",labelpad = 15)
plt.ylabel("Sales", labelpad = 15)
plt.legend(title="판매 연도")
```

```python
# 막대 그래프

labels = ["A", "B", "C", "D"]
values = [10,24,15,32]

# 수직 막대그래프
plt.bar(labels, values)
plt.show()
```

```python
# 수평 막대그래프
plt.barh(labels, values)
plt.show()
```

```python
# 여러 막대 그래프
labels = ["1분기", "2분기", "3분기", "4분기"]
men_means = [20,34,30,35]
women_means = [25,32,34,40]

x = np.arange(len(labels))

fig, ax = plt.subplots()
width = 0.4 # 막대 너비

ax.bar(x - width/2, men_means, width, label="남성")
ax.bar(x + width/2, women_means, width, label="여성")

ax.legend()
ax.set_xticks(x, labels) # 숫자 대신 문자열을 표시

plt.show()
```

#### 실습2. 막대 그래프 그리기

```python
categories = ['Category 1', 'Category 2', 'Category 3', 'Category 4', 'Category 5']
data = [20, 35, 15, 27, 45]

plt.bar(categories, data)

plt.title("Bar Chart")
plt.grid(linestyle=":")
plt.xlabel("Categories")
plt.ylabel("Values")

plt.grid(linestyle='-', alpha=0.5)

plt.xticks(rotation=45)
plt.yticks(rotation=45)

plt.ylim(0, 50) # y축을 0~50 범위로 고정

plt.show()
```

```python
# 히스토그램
# 랜덤 데이터 생성
data = np.random.randn(1000)

plt.hist(data, bins=30, edgecolor = "k")

plt.show()
```

#### 실습 3. 히스토그램 그리기

```python
# 임의로 1~6 사이의 주사위 눈 100개를 생성해서 각 숫자가 얼마나 자주 나왔는지 히스토그램을 그려보세요

dice = np.random.randint(1, 7, size=100)

cnt, bins, _ = plt.hist(dice, bins=np.arange(1,8)-0.5, edgecolor="black", rwidth=0.8)
print(bins)

plt.title("주사위 눈 빈도 히스토그램")
plt.xlabel("주사위 눈")
plt.ylabel("빈도")

plt.xticks([1,2,3,4,5,6])

plt.grid(axis = 'y', linestyle="--", alpha=0.5)

plt.show()
```

```python
# 산점도
x = np.linspace(2, 30, 20)
y = np.linspace(10, 40, 20)
y_noisy = y + np.random.uniform(-7, 7, 20)

plt.scatter(x, y_noisy)
plt.grid(linestyle = ":")

plt.show()
```

#### 실습4. 산점도 그리기

```python
# x값은 0부터 10까지 30개의 균등한 간격으로 구성된 수열을 만들고,
# y값은 y=3x+2+난수(정규분포, 평균 0, 표준편차 3)로 생성하여 산점도를 그려보세요

x = np.linspace(0, 10, 30)

noise = np.random.normal(0, 3, size=30)
y = 3 * x + 2 + noise

plt.scatter(x, y, color = "purple", marker = "s")

plt.title("노이즈가 추가된 산점도")
plt.xlabel("x 값")
plt.ylabel("y 값")

plt.grid(True, linestyle=':', alpha = 0.5)

plt.show()
```
