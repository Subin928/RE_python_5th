## 11. Selenium
- 웹 자동화를 위한 라이브러리

```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service

import time
```

```python
# Chrome 브라우저 실행
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

# 원하는 페이지로 이동
driver.get("https://www.google.com")
time.sleep(3)
driver.get("https://www.naver.com")
time.sleep(3)

# 페이지 이동
driver.back()
time.sleep(3)
driver.forward()
time.sleep(3)
driver.refresh()
time.sleep(3)

# 크기 변경
driver.set_window_size(1200, 800)
time.sleep(3)

# 최대화
driver.maximize_window()
time.sleep(3)

# 최소화
driver.minimize_window()
time.sleep(3)

# 창 배치
driver.set_window_position(50, 50)

# 스크린샷 저장
driver.save_screenshot("screenshot.png")

driver.quit()
```

```python
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
driver.get("https://www.selenium.dev/selenium/web/web-form.html")

# 웹 페이지의 요소 선택과 제어
text_input = driver.find_element(By.ID, "my-text-id")

time.sleep(3)

# 요소 제어
text_input.send_keys("Hello Selenium!")
time.sleep(3)
text_input.clear()
time.sleep(3)
text_input.send_keys("Coding")
time.sleep(3)

submit_btn = driver.find_element(By.TAG_NAME, "button")
submit_btn.click()

time.sleep(5)

driver.quit()
```

```python
# 구글 검색 예제

# 브라우저 실행
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

# Google 접속
driver.get("https://www.google.com")
time.sleep(3)

# 검색창 찾기 (name = 'q')
search_box = driver.find_element(By.NAME, "q")

# 기존 내용 지우기
search_box.clear()

# 검색어 입력 + Enter
search_box.send_keys("Python", Keys.RETURN)

time.sleep(15)

# 검색 결과 제목 5개 가져오기
titles = driver.find_elements(By.TAG_NAME, "h3")[:5]
for t in titles:
    print(t.text)

# 브라우저 종료
driver.quit()
```
