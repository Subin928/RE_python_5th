#### 조건 필터링

```python
data = {
    '이름' : ['홍길동', '김철수', '이영희', '박민수'],
    '점수' : [90, 70, 95, 95],
    '반' : [2, 1, 1, 2]
}

df = pd.DataFrame(data)
df
```

```python
mask = df["점수"] >= 80
filtered = df[mask]
filtered
```

```python
# 여러 조건 결합
# and : &
# or : |
# not : ~

# & 예시
df[(df["반"] == 1) & (df["점수"] >= 80)]
```

```python
# | 예시
df[(df["반"] == 2) | (df["점수"] >= 90)]
```

```python
# not 예시
df[~(df["점수"] >= 80)]
```

```python
# isin()
# 여러 값 중에 하나에 해당하는지 여부를 판단할 때 사용

# ex. 이름이 홍길동 이거나 박민수 이거나 한 경우
mask = (df["이름"] == "홍길동") | (df["이름"] == "박민수")
mask = df["이름"].isin(["홍길동", "박민수"])
df[mask]


mask2 = ~df["이름"].isin(["홍길동", "박민수"])
df[mask2]
```

```python
# reset_index
# 조건 필터링, 행 삭제 등을 통해 인덱스가 변경됐을 시
# reset_index를 사용 -> 인덱스가 초기화

mask = (df["반"] == 2) | (df["점수"] >= 90)
df2 = df[mask]
print(df2)

# 인덱스 리셋
# df2 = df2.reset_index()
# print(df2)

# 기존 인덱스 삭제
df2 = df2.reset_index(drop=True)
print(df2)
```

### 정렬
- 값 기준 정렬 : sort_values
- 인덱스 기준 정렬: sort_index

```python
# 값 기준 정렬 : sort_values
data = {
    '이름' : ['홍길동', '김철수', '이영희', '박민수'],
    '점수' : [90, 70, 85, 95],
    '반' : [2, 1, 1, 2]
}

df = pd.DataFrame(data)
df
```

```python
# 오름차순 정렬
df2 = df.sort_values("점수").reset_index(drop = True)
df2
```

```python
# 인덱스 기준 정렬 : sort_index
df_shuffled = df.sample(frac = 1, random_state = 42)
df_shuffled

# 행 인덱스 기준으로 정렬
df5 = df_shuffled.sort_index()  # 오름차순
df5

# 열 이름 기준 정렬 (알파벳순)
df6 = df.sort_index(axis = 1)
df6
```

### groupby

```python
df = pd.DataFrame({
    'team': ['A', 'A', 'B', 'B', 'B', 'C'],
    'name': ['Kim', 'Lee', 'Park', 'Choi', 'Jung', 'Han'],
    'score': [90, 85, 80, 70, 95, 88]
})
df
```

```python
# 단일 컬럼 기준 그룹화
grouped = df.groupby("team")

# 집계함수 적용
result_sum = grouped["score"].sum()
print(result_sum)

# 평균
result_mean = grouped["score"].mean()
print(result_mean)

# 개수
result_count = grouped["score"].count()
print(result_count)
```

```python
df2 = pd.DataFrame({
    'team': ['A', 'A', 'B', 'B', 'B', 'C'],
    'gender': ['M', 'F', 'F', 'M', 'M', 'F'],
    'score': [90, 85, 80, 70, 95, 88],
    'age' : [21, 22, 23, 25, 20, 27]
})
df2
```

```python
# 여러 컬럼 기준 그룹화
result = df2.groupby(["team", "gender"])["score"].mean()
result
```

```python
# as_index = False 옵션
# 그룹라벨이 인덱스로 설정됨
result = df2.groupby("team", as_index = False)["score"].sum()
result
```

```python
# 여러 집계 함수를 한 번에 적용 : agg()
result = df2.groupby("team", as_index = False)["score"].agg(["sum", "mean", "count"])
result
```

```python
# 그룹별로 여러 컬럼에 다른 집계 함수 적용
result = df2.groupby("team").agg({
    "score": "mean",
    "age": "max"
})
result
```

#### 실습 4. groupby 연습문제 (1)

```python
df = pd.DataFrame({
    'grade': [1, 2, 1, 2, 1, 3],
    'name': ['Kim', 'Lee', 'Park', 'Choi', 'Jung', 'Han'],
    'kor': [85, 78, 90, 92, 80, 75]
})
print(df)

# 1. 각 학년(grade)별 평균 국어 점수(kor)를 구하세요
df.groupby('grade')['kor'].mean()
```

```python
df = pd.DataFrame({
    'class': [1, 1, 1, 2, 2, 2],
    'subject': ['Math', 'Math', 'Eng', 'Math', 'Eng', 'Eng'],
    'score' : [80, 90, 85, 70, 95, 90]
})
print(df)

# 2. 아래 DataFrame에서 반(class)별, 과목(subject)별로 시험에 응시한 학생 수(count)와 평균 점수(avg)를 구하세요
df.groupby(['class', 'subject']).agg(count = ('score', 'count'), avg = ('score', 'mean'))
```

```python
df = pd.DataFrame({
    'region': ['Seoul', 'Seoul', 'Busan', 'Busan', 'Daegu'],
    'seller': ['A', 'B', 'A', 'B', 'A'],
    'sales': [100, 200, 150, 120, 130]
})
print(df)

# 3. 아래 DataFrame에서 지역(region)별 판매자(seller)별로 판매액(sales)의 합계와 최대값을 구하세요
df.groupby(['region', 'seller']).agg(sum_sales = ('sales', 'sum'), max_sales = ('sales', 'max'))
```

```python
df = pd.DataFrame({
    'team': ['A', 'A', 'B', 'B', 'A', 'B'],
    'position': ['FW', 'DF', 'FW', 'DF', 'DF', 'FW'],
    'score': [3, 2, None, 1, 4, 2]
})
print(df)

# # 4. 아래 DataFrame에서 팀(team)별, 포지션(position)별로 결측치(NaN)를 포함한 점수(score)의 평균을 구하세요
df.groupby(['team', 'position'])['score'].mean()
```

```python
df = pd.DataFrame({
    'dept': ['HR', 'HR', 'IT', 'IT', 'Sales', 'Sales'],
    'gender': ['M', 'F', 'F', 'M', 'F', 'F'],
    'salary': [3500, 3200, 4000, 4200, 3000, 3100]
})
print(df)

# 5. 아래 DataFrame에서 부서(dept)별로 성별(gender)별 인원 수와, 총 연봉(salary) 합계를 구하세요
df.groupby(['dept', 'gender']).agg(count = ('salary', 'count'), total_salary = ('salary', 'sum'))
```
