## 09. Folium
- 지도 시각화 라이브러리

```python
import folium
```

```python
# 기본 지도 생성
m=folium.Map(location=[37.611249848913126, 126.91726513654545], zoom_start = 18)

# 기본 마커
folium.Marker([37.611249848913126, 126.91726513654545], 
              popup="Subway",
              tooltip="구산역",
              icon=folium.Icon(color="black", icon="fa-solid fa-bus",icon_color="yellow", prefix="fa-solid")
              ).add_to(m)
m
```
