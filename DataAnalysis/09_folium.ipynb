## 09. Folium
- 지도 시각화 라이브러리

```python
import folium
```

```python
# 기본 지도 생성
import folium
m=folium.Map(location=[37.611249848913126, 126.91726513654545], zoom_start = 18)

# 기본 마커
folium.Marker([37.611249848913126, 126.91726513654545], 
              popup="Subway",
              tooltip="구산역",
              icon=folium.Icon(color="black", icon="fa-solid fa-train-subway",icon_color="yellow", prefix="fa-solid")
              ).add_to(m)

# 원형 마커
folium.CircleMarker(
    [37.6104,126.9128],
    radius = 100,
    color = "#adcdff",
    fill_color="#1c73ff",
    popuup = "CircleMarker",
    tooltip = "tooltip"
).add_to(m)


# 클릭한 곳에 마커 추가
m.add_child(folium.ClickForMarker(popup="내가 클릭한 곳"))

# 클릭한 곳의 위도와 경도 표시
m.add_child(folium.LatLngPopup())

m
```

```python
# 실습 1. 강의장 주변 장소 3개 표시하기
# 강의장 주변 맛집, 카페, 공원 등등 원하는 곳 3군데 이상 표시하기

# 기본 지도 생성
m = folium.Map(location=[37.61032097115849, 126.912939770091465], zoom_start = 18)

# 1. 딱다구리 커피
folium.Marker([37.61039962564502, 126.91269334542605], 
              popup="딱다구리 커피",
              tooltip="카페",
              icon=folium.Icon(color="white", icon="fa-solid fa-mug-saucer",icon_color="brown", prefix="fa-solid")
              ).add_to(m)

# 2. 구립구산동도서관마음
folium.Marker([37.60949892162893, 126.91307946590206], 
              popup="구립구산동도서관마을",
              tooltip="도서관",
              icon=folium.Icon(color="white", icon="fa-solid fa-book",icon_color="blue", prefix="fa-solid")
              ).add_to(m)

# 3. 거북골 근린공원
folium.Marker([37.61063179860596, 126.90701607229569], 
              popup="거북골근린공원",
              tooltip="공원",
              icon=folium.Icon(color="white", icon="fa-solid fa-tree",icon_color="green", prefix="fa-solid")
              ).add_to(m)
m
```

```python
# 실습 2. 택시 승차대 위치 표시하기
# "택시승차대 현황.csv"의 데이터로 택시 승차대 위치 표시하기!

import pandas as pd
import folium
df = pd.read_csv("./dataset/택시승차대 현황.csv", encoding='cp949')
# df.head()

taxi = df.loc[:, ["위치명", "위도", "경도"]]
latitude = taxi["위도"].mean()
longitude = taxi["경도"].mean()
m = folium.Map([latitude, longitude], zoom_start = 13)

for idx, row in taxi.iterrows():
    print(idx, row["위치명"], row["위도"], row["경도"])
    folium.Marker(
        [row["위도"], row["경도"]],
        tooltip = row["위치명"],
        icon = folium.Icon(color="orange", icon="fa-solid fa-taxi", prefix="fa-solid")
    ).add_to(m)

m
```
