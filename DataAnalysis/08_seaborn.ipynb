```python
# 스타일 설정

sns.set_style("ticks") # 스타일 테마 설정
sns.set_context("notebook") # 문맥에 따라 스타일 크기 조정
sns.set_palette("pastel")

plt.rcParams['font.family'] = 'Malgun Gothic'  # '맑은 고딕'이 설치되어 있을 경우
plt.rcParams['axes.unicode_minus'] = False     # 마이너스(-) 부호 깨짐 방지
```

```python
# 범주형 데이터 그래프
# countplot
plt.figure(figsize=(4, 3))
palette = sns.color_palette("pastel")
tips = sns.load_dataset("tips")
sns.countplot(data = tips, x="day", palette=palette)
plt.title("요일병 방문 횟수")
plt.show()
```

```python
# barplot
plt.figure(figsize=(4, 3))
palette = sns.color_palette("colorblind")
sns.barplot(x='day', y="tip",data=tips)
plt.title("요일별 평균 팁 금액")
plt.show()
```

```python
# displot()
# 히스토그램 + 커널 밀도 (kde)
# kde = True로 하면 kde 그래프 그려준다
plt.figure(figsize=(4, 3))
sns.displot(data=tips, x="total_bill", bins=20)
sns.displot(data=tips, x="total_bill", kde=True, bins=20)
plt.title("Total bill")
plt.show()
```

```python
# scatterplot()
# 산점도
plt.figure(figsize=(4, 3))
sns.scatterplot(data=tips, x="total_bill", y="tip", hue = "sex")
plt.title("Total Bill vs Tip (by Sex)")
plt.show()
```

```python
# heatmap()
# 상관계수행렬의 그래프
fig, ax = plt.subplots(figsize=(4, 3))

penguins = sns.load_dataset("penguins")
corr = penguins.corr(numeric_only=True)

sns.heatmap(corr, annot=True, fmt=".2f", cmap="inferno", ax=ax)
plt.show()

# 상관계수행렬
# - 상관계수: 두 변수 사이의 직선 관계가 얼마나 강한지를 숫자로 표현한 값
# - heatmap에서 색으로 표시되는 숫자가 상관계수
# 1: 양의 상관관계(하나가 증가하면 다른 것도 같은 비율로 증가)
# ex) 키와 몸무게
# -1: 음의 상관관계 (하나가 증가하면 다른 변수는 감소)
# ex) 공부시간과 실수 개수
```

```python
# jointplot()
sns.jointplot(data=penguins, x="flipper_length_mm", y="body_mass_g", kind="scatter", height = 4)
plt.show()
```

```python
# 종(species)별 히트맵 코드(실습)

# jointplot()
sns.jointplot(data=penguins, x="flipper_length_mm", y="body_mass_g", kind="scatter", hue="species")
plt.show()
```
